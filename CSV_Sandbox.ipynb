Using publically available Stats Canada data to calculate and plot the inflation rate and GDP growth over the past twenty years (2005 - 2024). 

Uncomment print statements for debugging.

## Define Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Calculating Inflation

```python
# We are only using the All-items row. 

CPU_frame = pd.read_csv("CPI.csv", sep=",", header=9, nrows=2, index_col=0)
all_inf_items_row = CPI_frame.loc['All-items']

# print(all_inf_items_row.to_string())
# print(CPI_frame.to_string())
```

```python
# Calculating inflation rate each year from the CPI. Drop the first year (2004) as it becomes nan.

inflation = ((all_inf_items_row.pct_change())*100).dropna().round(0).tolist()

# print(inflation)
```

## Import GDP

```python
# Retrieve row and drop the first colunm as to align size with inflation
GDP_frame = pd.read_csv("GDP.csv", sep=",", header=11, nrows=2, index_col=0)
GDP = GDP_frame.iloc[1].iloc[1:].tolist()
   
# print(GDP_frame)
# print(GDP)
```

## Plot

```python
fig_inf, ax_inf = plt.subplots()   

years = list(range(2004,2024))
every_other_year = [ year for year in years if year%2 == 0]

# print(years)
# print(every_other_year)
# print(inflation)
 
ax_inf.set_title("Inflation Over Time")
ax_inf.set_xlabel("Time")
ax_inf.set_ylabel("Inflation Rate")
ax_inf.plot(years, inflation, color="blue")
ax_inf.set_xticks(every_other_year)
plt.show()
```

```python

fig_GDP, ax_GDP = plt.subplots() 
ax_GDP.set_title("GDP Over Time")
ax_GDP.set_xlabel("Time")
ax_GDP.set_ylabel("GDP")
ax_GDP.plot(years, GDP, color="green")
ax_GDP.set_xticks(every_other_year)
plt.show()
```
